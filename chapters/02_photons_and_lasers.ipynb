# Photons: A Qualitative Overview
Light is famously made of *photons*, which act as the force carriers for the electromagnetic field in the standard model.  
A photon of light carries energy $E_\omega = \hbar \omega$,
where $\hbar$ is the reduced Planck's constant and
$\omega$ is the angular frequency of the light wave.

## Harmonic Oscillator Review
From Griffith's *Quantum Mechanics* Chapter 2.3,\
for a harmonic oscillator with potential $V(x) = \dfrac{1}{2} m \omega^2 x^2$,\
the energy levels are separated by $E = \hbar \omega$.

As it turns out, electric and magnetic fields can be quantized.\
As we've seen in the previous lecture, Maxwell's equations for electric and magnetic fields yield harmonic oscillator solutions,\
with eigenstates $|n \rangle$ with energy $E_n = \left(n + \dfrac{1}{2} \right) \hbar \omega $.

Crucially, the ground state $|0 \rangle$ is non-zero energy, and while the mean energy is zero, the variance in the ground state is non-zero, leading to measurable fluctuations in the field.

See Link 3: [Chapter 2.1 of Gerry and Knight *Introductory Quantum Optics*](https://dl.icdst.org/pdfs/files3/4dd1a65f65011e0c58716303d2d3f661.pdf) for an overview.

This is not a quantum optics class, so I will not emphasize the discrete nature of light,\
but we will need to remember the quantum nature of light when calculating shot noise or talking about squeezing later on.


Energy ladder for a harmonic oscillator from Griffith's QM.

Specifically, remember the following:
1. Maxwell's equations allow for any amount of energy in light waves.  Quantum mechanics limits us to discrete light particles called *photons*, each with energy $E = \hbar \omega$, where $\hbar$ is a very small number and $\omega$ is the light frequency.  
2. We can increase the *total energy* of light by stacking many photons: $E_\mathrm{total} = n \hbar \omega$, where $n$ is the number of photons available.
3. Quantum effects typically start becoming important when $n = 0$ or $n = 1$.  Otherwise light can be considered to behave *semiclassically*.  It is in the semiclassical limit we will mostly operate.  

# Lasers

Here we will review some important aspects of lasers as presented in Chapter 1 of *Lasers* by Seigman.

A *laser* is a device that generates or amplifies **coherent** light at the same frequency.

Any laser must be comprised of three elements:
1. A *laser medium* consisting of an appropriate collection of atoms, molecules, ions, or in some instances a semiconducting crystal;
2. A *pumping process* to excite these atoms (molecules, etc.) into higher quantum-mechanical energy levels;
3. *Optical feedback elements* that allow a beam of radiation to either pass once through the laser medium (as in a laser amplifier) or bounce back and forth repeatedly through the laser medium (as in a laser oscillator).

## Stimulated Emission
The core mechanism behind lasers is *stimulated emission*. 
Stimulated emission is the process of using one photon to generate an exact copy of itself from an excited (or *pumped*) atom.
The copy has the same frequency and phase
